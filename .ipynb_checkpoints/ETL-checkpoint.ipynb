```python
import pandas as pd
from sqlalchemy import create_engine
from key import key
```

```python
wine_file = "./Resources/winemag-data-130k-v2.csv"
wine_df = pd.read_csv(wine_file)
wine_df.columns
```

```python
wine_df.head()
```

```python
new_wine_df=wine_df[['country','designation','points','price','province','region_1','variety','winery']].copy()
wine_data=new_wine_df.dropna()
wine_data.head()
new_wine_data=wine_data.rename(columns={"country":"country","designation":"designation", "points":"points",
                            "price":"price","province":"province","region_1":"region",
                                       "variety":"variety", "winery":"winery"})

new_wine_data['id'] = range(1, len(new_wine_data) + 1)
new_wine_data=new_wine_data[['id','country','designation','points','price','province','region','variety','winery']]

new_wine_data.to_csv('wine_data.csv', index=False)
```

```python
new_wine_data.head()
```

```python
beer_file = "./Resources/beer_reviews.csv"
beer_df = pd.read_csv(beer_file)
beer_df.head()
```

```python
new_beer_df=beer_df[['brewery_name','review_overall','beer_style','beer_name','beer_abv']]
new_beer_df
```

```python
beer_data=new_beer_df.drop_duplicates()
beer_data=beer_data.dropna()
new_beer_data=beer_data.rename(columns={"brewery_name":"brewery","review_overall":"overall_review", "beer_style":"beer_style",
                            "beer_name":"beer_name","beer_abv":"abv"})
new_beer_data['id'] = range(1, len(new_beer_data) + 1)
new_beer_data=new_beer_data[['id','brewery','overall_review','beer_style','beer_name','abv']]
new_beer_data.head()
```

```python
new_beer_data.to_csv('beer_data.csv', index=False)
```

```python
rds_connection_string = (f"postgres:{key}@localhost:5432/etl")
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_wine_data.to_sql(name='wine', con=engine, if_exists='append', index=False)
```

```python
new_beer_data.to_sql(name='beer', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from wine', con=engine).head()
```

```python
pd.read_sql_query('select * from beer', con=engine).head()
```
